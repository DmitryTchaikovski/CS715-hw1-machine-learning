```python
# import libraries and checking versions
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import matplotlib
import sklearn 
from sklearn import datasets
```

```python
# print versions of each library to verify installation
print("NumPy version:", np.__version__)
print("Pandas version:", pd.__version__)
print("Matplotlib version:", matplotlib.__version__)
print("Seaborn version:", sns.__version__)
print("SciPy version:", scipy.__version__)
print("Scikit-learn version:", sklearn.__version__) 
```

```python
# create a simple test plot using Matplotlib
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Test Plot")
plt.show()
```

```python
# loading a sample dataset from Scikit-learn and printing its description
iris = datasets.load_iris()
print("\nIris Dataset Description:")
print(iris['DESCR'])
```

```python
# creating a Seaborn plot from the Iris dataset
sns.scatterplot(x=iris.data[:, 0], y=iris.data[:, 1], hue=iris.target)
plt.title("Seaborn Test Plot - Iris Dataset")
plt.show()
```
